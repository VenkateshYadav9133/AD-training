```python
import speech_recognition as sr
import pyttsx3
import pywhatkit
import datetime
import wikipedia
import pyjokes
import pychrome
import subprocess
import urllib.parse
import webbrowser
import pyautogui
import openai
listener = sr.Recognizer()
engine = pyttsx3.init()
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)


def talk(text):
     engine.say(text)
     engine.runAndWait()


def open_youtube():
    webbrowser.open("https://www.youtube.com/")
    talk("Opening YouTube...")

def open_chrome():
    subprocess.Popen(['C:\\Program Files\\Google\\Chrome\\Application\\chrome.exe'])

def search(query):
    search_url = f"https://www.google.com/search?q={query}"
    webbrowser.open(search_url)
    talk(f"Searching for {query}...")


def take_command():
    command = None
    try:
        with sr.Microphone() as source:
            print('listening...')
            voice = listener.listen(source)
            command = listener.recognize_google(voice)
            command = command.lower()
            if 'edith' in command:
                command = command.replace('edith', '').strip()
                print(command)
    except Exception as e:
        print(f"Error: {e}")
    return command
def run_edith():
    command = take_command()
    if command is None:
        return
    print(command)
    if 'play' in command:
        song = command.replace('play', '')
        talk('playing ' + song)
        pywhatkit.playonyt(song)
    elif 'time' in command:
        time = datetime.datetime.now().strftime('%I:%M %p')
        talk('Current time is ' + time)
    elif 'who is' in command:
        person = command.replace('who is', '')
        info = wikipedia.summary(person, 1)
        print(info)
        talk(info)
    elif "date" in command:
        today_date = datetime.datetime.now ().strftime ("%A, %B %d, %Y")
        talk(f"Today's date is {today_date}")
    elif 'are you single' in command:
        talk('I am in a relationship with wifi')
    elif 'joke' in command:
        talk(pyjokes.get_joke())
    elif "open notepad" in command.lower():
        subprocess.Popen (['notepad.exe'])
    elif "open calculator" in command.lower():
        subprocess.Popen (['calc.exe'])
    elif "youtube" in command:
        open_youtube ()
        talk("Opening YouTube...")
    elif "chrome" in command:
        open_chrome ()
        talk("Opening Chrome...")
    elif 'search' in command:
        query = command.replace('search', '').strip()
        search(query)
    else:
        talk('I love you bro.')
while True:
    run_edith()
```
